```python
# Using this notebook to clean some of the data, and merge the two cameras.
```

```python
import pandas as pd
```

```python
cam1 = pd.read_csv('ai-data.csv')
cam2 = pd.read_csv('ai-data-2.csv')
```

```python
cam1.head()
```

```python
cam2.head()
```

```python
cam2['datetime'] = pd.to_datetime(cam2['datetime'])
cam2.head()
```

```python
cam2['hour'] = cam2['datetime'].dt.hour
cam2['date'] = cam2['datetime'].dt.date
```

```python
cam2.head()
```

```python
cam1 = cam1[['camera', 'date', 'hour', 'zone', 'person', 'car', 'bicycle', 'motorcycle', 'truck', 'dog']]
cam2 = cam2[['camera', 'date', 'hour', 'zone', 'person', 'car', 'bicycle', 'motorcycle', 'truck', 'dog']]
```

```python
cam1.head()
```

```python
cam2
```

```python
df = pd.concat([cam1, cam2])
df.reset_index(drop=True, inplace=True)
df
```

```python
df.to_csv('merged-ai-data.csv')
```
