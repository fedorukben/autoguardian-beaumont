```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pprint
```

```python
pp = pprint.PrettyPrinter(indent=4)
```

```python
df = pd.read_csv('merged-ai-data.csv')[['camera', 'date', 'hour', 'zone', 'person', 'car', 'bicycle', 'motorcycle', 
                                       'truck', 'dog']]
df['date'] = pd.to_datetime(df['date'])
df
```

```python
df.dtypes
```

```python
lst = df[df['camera'] == 1]['zone'].tolist()
lst.extend((df[df['camera'] == 2]['zone'] + 3).tolist())
df['camzone'] = lst
df
```

```python
czs = df['camzone'].unique()
```

```python
for cz in czs:
    pp.pprint(df[df['camzone'] == cz].groupby('date').mean()[['camera', 'zone', 'person', 'car']])
```
